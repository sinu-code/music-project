```python

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
sns.histplot(df['mfcc5'], kde=True, bins=30, color='C2')
plt.title('mfcc5 Distribution')
plt.xlabel('mfcc5')

plt.subplot(1, 3, 2)
sns.boxplot(x=df['mfcc5'], color='C3')
plt.title('mfcc5 Boxplot')

plt.subplot(1, 3, 3)
plt.plot(df['mfcc5'].values, marker='.', linestyle='none', alpha=0.6)
plt.title('mfcc5 (by index)')
plt.xlabel('Index')
plt.ylabel('mfcc5')

plt.tight_layout()
plt.show()
```
